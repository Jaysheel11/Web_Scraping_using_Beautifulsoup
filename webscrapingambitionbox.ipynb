```python
import pandas as pd
from bs4 import BeautifulSoup
import requests




```

```python
final = pd.DataFrame()
webpage = []

for i in range(0, 333):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
    webpage.append(requests.get(f"https://www.ambitionbox.com/list-of-companies?sort_by=popularity&page={i}", headers=headers).text)

```

```python
for page_content in webpage:
    soup = BeautifulSoup(page_content, 'html.parser')
    company = soup.find_all('div', class_='company-content-wrapper')
    name = []
    rating = []
    reviews = []
    ctype = []
    hq = []
    old = []
    employees = []

    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('p', class_='rating').text.strip())
        reviews.append(i.find('a', class_='review-count').text.strip())
        
        info_entities = i.find_all('p', class_='infoEntity')
        if len(info_entities) >= 4:
            ctype.append(info_entities[0].text.strip())
            hq.append(info_entities[1].text.strip())
            old.append(info_entities[2].text.strip())
            employees.append(info_entities[3].text.strip())
        else:
            ctype.append('')
            hq.append('')
            old.append('')
            employees.append('')

    d = {
        'name': name,
        'rating': rating,
        'reviews': reviews,
        'type': ctype,
        'hq': hq,
        'old': old,
        'employees': employees
    }

    df = pd.DataFrame(d)
    final = final.append(df)
```
